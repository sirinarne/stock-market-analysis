# Financial Metrics

## Objective

Read the historical stock price dataset and calculate financial metrics including:

- Daily dollar change
- Daily percentage return
- 7-Day Moving Average
- 20-Day Moving Average
- 50-Day Moving Average
- Rolling Volatility

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/stock_prices.csv")
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
df = df.sort_values(
    by=["Ticker", "Date"]
)
df
```

```python
# Daily Dollar Change 
# today's close - yesterday's close 

df["Dollar Change"] = (
    df.groupby("Ticker")["Close"]
      .diff()
)
df
```

```python
# Daily Percentage Return 
# (Today's Close - Yesterday's Close)/ (Yesterday's Close)
df["Daily Return"] = (
    df.groupby("Ticker")["Close"]
      .pct_change()
)
df["Daily Return (%)"] = (
    df["Daily Return"] * 100
)
df["Daily Return (%)"] = (
    df["Daily Return (%)"].round(2)
)

df
```

```python
# Seven - Day Moving Average - smooths daily price fluctuations
# Takes the previous 7 closing prices and calculates their average


df["7-Day MA"] = (
    df.groupby("Ticker")["Close"]
      .transform(
          lambda x: x.rolling(7).mean()
      )
)

```

```python
# Twenty - Day Moving Average
df["20-Day MA"] = (
    df.groupby("Ticker")["Close"]
      .transform(
          lambda x: x.rolling(20).mean()
      )
)
```

```python
# Fifty - Day Moving Average

df["50-Day MA"] = (
    df.groupby("Ticker")["Close"]
      .transform(
          lambda x: x.rolling(50).mean()
      )
)
```

```python
# Rolling Volatility - 20 day 
# Measures how much a stock's price fluctuates - informs us of how much daily returns vary over time
# rolling SD of daily returns (how much a stock's price changed from one day to next) over a 20 day window 
# returns close together = low volatility 
# returns swing dramatically = high volatility

df["20-Day Volatility"] = (
    df.groupby("Ticker")["Daily Return"]
      .transform(lambda x: x.rolling(20).std())
)
```

```python
df.to_csv(
    "../data/processed/stock_metrics.csv",
    index=False
)
```
