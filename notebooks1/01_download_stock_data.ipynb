## Downloading Daily Stock Market Data

```python
import pandas as pd
from datetime import datetime
import yfinance as yf
```

```python
stocks =[
    "AAPL",
    "MSFT",
    "NVDA",
    "AMZN",
    "JPM",
    "SPY"
]
```

```python
today = datetime.today().strftime("%Y-%m-%d")
today
```

```python
rows =[]
```

```python
for ticker in stocks:
    stock = yf.Ticker(ticker)
    history = stock.history(period = "1d")

    if history.empty:
        print(f"No data available for {ticker}")
        continue

    open_price = history["Open"].iloc[-1] # price when market opened that day
    high_price = history["High"].iloc[-1] # highest price stock reached that day
    low_price = history["Low"].iloc[-1] #lowest price
    close_price = history["Close"].iloc[-1] # price when market closed that day
    volume = history["Volume"].iloc[-1] # number of shares traded that day

    rows.append({
    "Date": today,
    "Ticker": ticker,
    "Open": round(open_price, 2),
    "High": round(high_price, 2),
    "Low": round(low_price, 2),
    "Close": round(close_price, 2),
    "Volume": int(volume)
})
    
df = pd.DataFrame(rows)
print(df)
```

## Building a Historical Stock Database

```python
import os
```

```python
csv_file = "../data/raw/stock_prices.csv"
```

```python
os.path.exists(csv_file)
```

```python
if os.path.exists(csv_file):
    existing_df = pd.read_csv(csv_file)
    print("Existing dataset loaded.")
else:
    existing_df = pd.DataFrame()
    print("No existing dataset found.")
```

```python
updated_df = pd.concat([existing_df, df], ignore_index=True)
```

```python
updated_df = updated_df.drop_duplicates(
    subset=["Date", "Ticker"],
    keep="last"
)
```

```python
updated_df = updated_df.sort_values(
    by=["Date", "Ticker"]
)
```

```python
updated_df = updated_df.reset_index(drop=True)
updated_df.to_csv(csv_file, index=False)
```

```python
print(f"Dataset saved to: {csv_file}")
```
